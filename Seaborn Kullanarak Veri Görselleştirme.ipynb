## Seaborn Plotlama İşlevleri

İstatistiksel İlişkileri Görselleştirme:
Bir veri kümesinin değişkenleri arasındaki ilişkileri anlama süreci ve bu ilişkilerin sırayla diğer değişkenlere nasıl bağlı olduğu, istatistiksel analiz olarak bilinir. Şimdi bunun için gereken işlevlere daha derin bir göz atalım:

**relplot():**
Bu, İstatistiksel İlişkileri Görselleştirmek için diğer iki eksen işlevini kullanan bir şekil düzeyinde işlevdir:

_scatterplot()_

_lineplot()_

Bu işlevler relplot()'un tür parametresi kullanılarak belirtilebilir. Bu parametrenin verilmesi durumunda, scatterplot() olan varsayılanı alır. Kodunuzu yazmaya başlamadan önce, gerekli kitaplıkları aşağıdaki gibi içe aktardığınızdan emin olun:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

Şimdi küçük bir örneğe bakalım:

```python
f = sns.load_dataset("flights") 
sns.relplot(x="passengers", y="month", data=f);
```

Gördüğünüz gibi, noktalar 2 boyutlu olarak çizilmiştir. Ancak, "hue" kullanarak başka bir boyut ekleyebilirsiniz. Aynı örneğe bir göz atalım:

```python
f = sns.load_dataset("flights") 
sns.relplot(x="passengers", y="month", hue="year", data=f);
```

Bununla birlikte, renkler, stiller, boyut vb. Gibi deneyebileceğiniz daha birçok özelleştirme var. Aşağıdaki örnekte rengi nasıl değiştirebileceğinizi görebilirsiniz:

```python
sns.set(style="darkgrid")
f = sns.load_dataset("flights")
sns.relplot(x="passengers", y="month", hue="year",palette="ch:r=-.5,l=.75", data=f);
```

## lineplot():

Bu işlev, verileriniz için kesintisiz bir çizgi çizmenize izin verecektir. Bu işlevi, "kind" parametresini aşağıdaki gibi değiştirerek kullanabilirsiniz:

```python
a=pd.DataFrame({'Day':[1,2,3,4,5,6,7],'Grocery':[30,80,45,23,51,46,76],'Clothes':[13,40,34,23,54,67,98],'Utensils':[12,32,27,56,87,54,34]},index=[1,2,3,4,5,6,7]) 
g = sns.relplot(x="Day", y="Clothes", kind="line", data=a) 
g.fig.autofmt_xdate()
```

Çizgi grafiği için varsayılan, x'in bir fonksiyonu olarak y'dir. Ancak, isterseniz değiştirilebilir. Daha fazla deneyebileceğiniz daha birçok seçenek var.
Şimdi kategorik verileri nasıl çizeceğimize bir göz atalım.

## Kategorik Verilerle Çizim Yapma:

Bu yaklaşım, ana değişkenimiz ayrık gruplara (kategorik) bölündüğünde ortaya çıkar. Bu, catplot() işlevi kullanılarak elde edilebilir.

## catplot ():

Bu, relplot () gibi şekil düzeyinde bir işlevdir. Üç eksen seviyesi fonksiyonu ailesi ile karakterize edilebilir, yani:

1- Scatterplots - Bunlar stripplot(), swarmplot().

2- Distribution Plots - Bunlar boxplot(), violinplot(), boxenplot().

3- Estimateplots - Bunlar pointplot(), barplot(), countplot().

Şimdi bunu göstermek için birkaç örnek verelim:

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
a = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", data=a);
```

Gördüğünüz gibi, yukarıdaki örnekte "kind" parametresini ayarlamadım. Bu nedenle grafiği varsayılan dağılım grafiği olarak döndürdü. Grafiği gerektiği gibi değiştirmek için eksen seviyesi işlevlerinden herhangi birini belirtebilirsiniz. Buna da bir örnek verelim:

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
a = sns.load_dataset("tips")
sns.catplot(x="day", y="total_bill", kind="violin", data=a);
```

Yukarıdaki çıktı, ipuçları veri kümesi için violinplot gösterir. Şimdi bir veri kümesinin dağılımını nasıl görselleştireceğimizi bulmaya çalışalım.

## Bir veri kümesinin dağıtımını görselleştirme:

Bu temelde, tek değişkenli veya iki değişkenli olmak için bağlamla veri kümelerini anlamakla ilgilidir. Buna başlamadan önce aşağıdakileri içe aktarmanız yeterlidir:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
sns.set(color_codes=True)
```

Bu yapıldıktan sonra, tek değişkenli ve iki değişkenli dağılımları çizmeye devam edebilirsiniz.

## Tek değişkenli dağılımları çizme:

Bunların grafiğini çizmek için distplot() işlevini aşağıdaki gibi kullanabilirsiniz:

```python
a = np.random.normal(loc=5,size=100,scale=2) 
sns.distplot(a);
```

Yukarıdaki örnekte görebileceğiniz gibi, değerleri distplot kullanılarak normal() fonksiyonu tarafından üretilen a değişkeni için bir grafik çizdik.

## İki değişkenli dağılımları çizme:

Bu, olası bir olayla sonuçlanan iki rastgele bağımsız değişkeniniz olduğunda ortaya çıkar. Bu tür grafikleri çizmek için en iyi işlev, jointplot(). Şimdi, jointplot() kullanarak iki değişkenli bir grafik çizelim.

```python
x=pd.DataFrame({'Day':[1,2,3,4,5,6,7],'Grocery':[30,80,45,23,51,46,76],'Clothes':[13,40,34,23,54,67,98],'Utensils':[12,32,27,56,87,54,34]},index=[1,2,3,4,5,6,7])
y=pd.DataFrame({'Day':[8,9,10,11,12,13,14],'Grocery':[30,80,45,23,51,46,76],'Clothes':[13,40,34,23,54,67,98],'Utensils':[12,32,27,56,87,54,34]},index=[8,9,10,11,12,13,14])
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
with sns.axes_style("white"):sns.jointplot(x=x, y=y, kind="kde", color="b");
```

Artık Python Seaborn'daki çeşitli işlevleri anladığınıza göre, yapılandırılmış çoklu plan ızgaraları oluşturmaya geçelim.

## Çok Alanlı Izgaralar:

Python Seaborn, birden fazla ızgarayı yan yana çizmenize olanak tanır. Bunlar temelde, aralarındaki karşılaştırmaya yardımcı olmak için aynı ölçek ve eksenler kullanılarak çizilen grafikler veya grafiklerdir. Bu da programcının grafikler arasında hızla ayrım yapmasına ve büyük miktarda bilgi edinmesine yardımcı olur.       

Bu grafikleri çizmek için aşağıdaki facetgrid() işlevi örneğini düşünün.

```python
sns.set(style="darkgrid") 
a = sns.load_dataset("iris") 
b = sns.FacetGrid(a, col="species") 
b.map(plt.hist, "sepal_length");
```

Yukarıdaki çıktı, Öğle ve Akşam Yemeği sırasında verilen ipuçları arasındaki karşılaştırmayı açıkça göstermektedir. Karşılaştırmak için bir çift değişkeniniz olduğunda, PairGrid işlevini kullanarak da çizim yapabilirsiniz. Aşağıdaki örneği düşünün.

```python
sns.set(style="ticks") 
a = sns.load_dataset("flights") 
b = sns.PairGrid(a) 
b.map(plt.scatter);
```

Gördüğünüz gibi, yukarıdaki çıktı, yıl ve yolcu sayısını farklı şekillerde açıkça karşılaştırıyor.

Seaborn ayrıca estetikle ilgili özelleştirmelere de izin veriyor.

## Python Seaborn Figür Estetiği:

Tartışacağım ilk işlev set(). Bu işlevin 'style' parametresini daha önce kullanıyordum. Bu parametre temelde deniz doğusu temaları ile ilgilidir. Şu anda darkgrid, ticks, whitegrid, white and dark olmak üzere beş tane mevcuttur.

Beyaz temayı gösteren aşağıdaki örneği düşünün.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
a = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=a);
```

Yukarıdaki çıktıda temanın beyaza dönüştüğünü fark edebilirsiniz. Diğer temaları da kullanarak onları daha fazla keşfedebilirsiniz. Önceki çıktıda fark ettiyseniz, grafiğin her yerinde eksenler vardı. Ancak bu, despine() işlevi kullanılarak da özelleştirilebilir. Aşağıdaki örneğe bakın.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
a = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=a);
sns.despine(offset=10, trim=True);
```

## Python Seaborn Renk Paletleri:

Renk, temelde insan gözüne başka herhangi bir özelliğin ötesinde yaklaşan özelliktir. Seaborn, color_palette(), hls_palette(), husl_palette(), vb. Gibi çeşitli işlevleri kullanarak renklerle oynamanıza izin verir. Seaborn'da şu anda mevcut olan renklere bir göz atın.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
 
presentcolors = sns.color_palette()
sns.palplot(presentcolors)
```

Yukarıdaki görüntü, seaborn'da bulunan renkleri göstermektedir. Palplot() işlevini kullanarak yaptım. Daha derin varyasyonlar için hls_palette(), husl_palette(), vb. Kullanabilirsiniz.
